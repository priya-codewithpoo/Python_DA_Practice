2. Mean, median, std using NumPy

```python
numbers = np.array([20, 30, 40, 50, 60])

num_mean = np.mean(numbers)
num_median = np.median(numbers)
num_std = np.std(numbers).astype(int)

print('num_mean:', num_mean)
print('num_median:', num_median)
print('num_std:', num_std)
```

3. Replace values in NumPy array

```python
arr = np.array([10, 15, 20, 25, 30])

arr[arr > 20] = 100
print(arr)
```

4. NumPy array filtering

```python
arr = np.array([3, 7, 10, 15, 20, 25, 30])

filtered = arr[arr > 20]
filtered
  
```

```python
filtered = arr > 20  # boolean
filtered
```

5. Reshape NumPy array
