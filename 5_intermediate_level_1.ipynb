```python
import datasets
from datasets import load_dataset
dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()

```

## Aggregations & Grouping

1. Count jobs by job title

```python
df_job_count = df.groupby('job_title_short').size().reset_index(name='job_count').sort_values(by = 'job_count' , ascending = False)
df_job_count
```

2. Average salary by country

```python
avg_salary_by_country = df[df['salary_year_avg'].notna()]
avg_salary_by_country =(avg_salary_by_country.groupby('job_country')['salary_year_avg']
                        .mean()
                        .astype(int)     
                        .reset_index(name='avg_salary')
                        .sort_values(by = 'avg_salary',ascending = False)
)
avg_salary_by_country


```

3. Top 5 countries by job count

```python
top_5_countries =(df.groupby('job_country')['job_title_short'].size()
                  .reset_index(name = 'job_count')
                  .sort_values(by = 'job_count', ascending=False)
                  .head(5)
                  )
top_5_countries
```

4. Highest paying job title

```python
highest_paying_job =(df.groupby('job_title_short')['salary_year_avg']
                     .max()
                     .reset_index(name= 'highest_salary')
                     .sort_values(by = 'highest_salary', ascending = False)
                     .head(1)
                     )
highest_paying_job
```

5. Average salary by work type (remote / onsite)

```python
work_type_salary = (df.groupby('job_work_from_home')['salary_year_avg']
                   .mean()
                   .astype(int)
                   .reset_index(name = 'avg_salary')
                   .sort_values(by = 'avg_salary',ascending=False)
                   )
work_type_salary
```

6. Job count by company

```python
job_by_company =(df.groupby('company_name')['job_title_short']
                 .size()
                 .reset_index(name = 'job_count')
                 .sort_values(by = 'job_count',ascending = False)
                 )
job_by_company
```

7. Average salary by job schedule

```python
avg_salary_by_shedule = df[df['salary_year_avg'].notna()]
avg_salary_by_shedule =(avg_salary_by_shedule.groupby('job_schedule_type')['salary_year_avg']
                 .mean()
                 .round()
                 .reset_index(name = 'avg_salary')
                 .sort_values(by = 'avg_salary',ascending = False)
                 )
avg_salary_by_shedule
```

8. % of remote jobs

```python

percent_remote = (df['job_work_from_home'] == True).sum() / len(df) * 100
percent_remote = round(percent_remote, 2)
percent_remote
```

9. Compare degree vs non-degree salary

```python
degree_salary = (df.groupby('job_no_degree_mention')['salary_year_avg'].mean()
                .astype(int)
                .reset_index(name = 'avg_salary')
                .sort_values(by = 'avg_salary', ascending = False))
degree_salary
```

10. Top 10 highest paying companies

```python
top_10_companies = (df.groupby('company_name')['salary_year_avg'].mean()
                    .round()
                    .reset_index(name='salary')
                    .sort_values(by = 'salary',ascending=False)
                    .head(10))
top_10_companies
```
