```python
import datasets
from datasets import load_dataset
import ast
dataset = load_dataset('lukebarousse/data_jobs')

df = dataset['train'].to_pandas()
```

## Sorting & Ranking

 11. Top 10 highest salaries

```python
top_ten_salary = df[['job_title_short', 'salary_year_avg']].sort_values(by='salary_year_avg', ascending=False).head(10)
top_ten_salary
```

12. Top 10 lowest salaries

```python
top_ten_salary = df[['job_title_short', 'salary_year_avg']].sort_values(by='salary_year_avg').head(10)
top_ten_salary
```

13. Jobs sorted by post date

```python
import pandas as pd
from datetime import datetime

df['job_posted_date'] = pd.to_datetime(df['job_posted_date']).dt.date
sorted_date =df[['job_posted_date','job_title']].sort_values(by = 'job_posted_date',ascending=False)
sorted_date
```

14. Most recent job posting

```python
recent_job = [df[['job_title_short','job_posted_date']].max()]
recent_job
```

15. Companies posting the highest number of jobs

```python
companies_with_more_job = (df.groupby('company_name')
                           .size()
                           .reset_index(name='job_count')
                           .sort_values(by = 'job_count',ascending=False)
                           .head()
)
companies_with_more_job
```

16. Highest paying remote jobs

```python
# filter for remote jobs
remote_jobs = df[df['job_work_from_home'] == True] 
remote_jobs = remote_jobs.sort_values(by='salary_year_avg', ascending=False).head()

remote_jobs[['job_title_short', 'salary_year_avg']]
```

17. Highest paid job in India

```python
jobs_in_india = df[df['job_country'] == 'India']
jobs_in_india = jobs_in_india.sort_values(by='salary_year_avg', ascending=False).head(1)

jobs_in_india[['job_title_short', 'salary_year_avg','job_country']]
```

18. Highest paid job type

```python

df_sal = df[df['salary_year_avg'].notna()]
df_sal =(df_sal.groupby('job_schedule_type')['salary_year_avg']
         .mean()
         .reset_index(name='salary')
         .sort_values(by = 'salary',ascending=False).max()
)
df_sal
```

19. Highest salary per country

```python
highest_paid_salary = df[df['salary_year_avg'].notna()]
highest_paid_salary =(highest_paid_salary.groupby('job_country')['salary_year_avg']
                      .max()
                      .reset_index(name='highest_salary')
                      .sort_values(by='highest_salary',ascending =False)
)
highest_paid_salary 
```

20. Top 5 skills demanded

```python
import ast
import pandas as pd

# Convert string representation of lists into actual lists
df['job_skills'] = df['job_skills'].apply(lambda x: ast.literal_eval(x) if pd.notna(x) else x)

# Explode the skills into separate rows
skills_exploded = df.explode('job_skills').reset_index(drop=True)

skills_exploded['job_skills'].value_counts().sort_values(ascending=False).head(5)
```
