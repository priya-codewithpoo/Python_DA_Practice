1. Write a function to find the average of a list

```python
def avg(numbers):
    return sum(numbers) / len(numbers)


# Example usage:
nums = [10, 20, 30, 40]
print(avg(nums))
```

2. Function that finds max and min from a list

```python
def find_max_min(numbers):
    if len(numbers) == 0:
        return "List is empty"

    max_val = numbers[0]
    min_val = numbers[0]

    for num in numbers:
        if num > max_val:
            max_val = num
        if num < min_val:
            min_val = num

    return max_val, min_val


# Example
nums = [12, 4, 7, 99, 1]
mx, mn = find_max_min(nums)

print("Max:", mx)
print("Min:", mn)
```

3. Function to count even numbers

```python
def count_even(numbers):
    count = 0
    for num in numbers:
        if num % 2 == 0:
            count += 1
    return count

my_list = [1, 2, 3, 4, 6, 9]
print(count_even(my_list))

```

4. Function to filter numbers above a threshold

```python
def filter_above(numbers, threshold):
    result = []
    for num in numbers:
        if num > threshold:
            result.append(num)
    return result

nums = [5, 10, 3, 20, 7]
output = filter_above(nums, 8)

print("Result:", output)


```

5. Function to clean strings

```python
import re   # for removing symbols

def clean_string(text):
        text = text.lower()
    
     # Remove special characters using regex
        text = re.sub(r'[^a-z0-9\s]', '', text)
    
    # Remove extra spaces
        text = text.strip()
    
        return text

word = "  Hello!! Python @2025  "
print(clean_string(word))


 
```
